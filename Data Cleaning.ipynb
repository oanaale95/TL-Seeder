```python
def get_row_dict(acquisition_id, acquiring_object_id, acquired_object_id, term_code, price_amount, price_currency_code,
               acquired_at, source_url, source_description, created_at, updated_at, full_name, subject, object_id, aquired):
    
#     global df_acq_people
    
    acq_people_row_dict = {}
    
    acq_people_row_dict['acquisition_id'] = acquisition_id
    acq_people_row_dict['acquiring_object_id']= acquiring_object_id
    acq_people_row_dict['acquired_object_id']= acquired_object_id
    acq_people_row_dict['term_code']= term_code
    acq_people_row_dict['price_amount']= price_amount
    acq_people_row_dict['price_currency_code']=price_currency_code
    acq_people_row_dict['acquired_at']=acquired_at
    acq_people_row_dict['source_url']= source_url
    acq_people_row_dict['source_description']= source_description
    acq_people_row_dict['created_at']=created_at
    acq_people_row_dict['updated_at']=updated_at
    acq_people_row_dict['full_name']=full_name
    acq_people_row_dict['subject']=subject
    acq_people_row_dict['object_id']=object_id
    acq_people_row_dict['aquired']=aquired
    
#     df_acq_people= df_acq_people.append(acq_people_row_dict,ignore_index=True)
    return acq_people_row_dict
```

```python
rows_list = []
for row_acq in range(df_acquisitions.shape[0]):
    for row_people in range(df_people.shape[0]):
        # if the object id refers to acquiring_object_id => acquired = False
        if df_acquisitions.loc[row_acq,'acquiring_object_id'] == df_people.loc[row_people ,'relationship_object_id']:
            rows_list.append(
                get_row_dict(
                    df_acquisitions.loc[row_acq, 'acquisition_id'],
                    df_acquisitions.loc[row_acq, 'acquiring_object_id'],
                    df_acquisitions.loc[row_acq, 'acquired_object_id'],
                    df_acquisitions.loc[row_acq, 'term_code'],
                    df_acquisitions.loc[row_acq, 'price_amount'],
                    df_acquisitions.loc[row_acq, 'price_currency_code'],
                    df_acquisitions.loc[row_acq, 'acquired_at'],
                    df_acquisitions.loc[row_acq, 'source_url'],
                    df_acquisitions.loc[row_acq, 'source_description'],
                    df_acquisitions.loc[row_acq, 'created_at'],
                    df_acquisitions.loc[row_acq, 'updated_at'],
                    df_people.loc[row_people, 'full_name'],
                    df_people.loc[row_people, 'subject'],
                    df_people.loc[row_people, 'object_id'],  
                    False             
                )
            )
            
           # print('acquring ' + str(row_acq) + ' ' + str(row_people))
            
        # if the object id refers to acquired_object_id => acquired = True
        elif df_acquisitions.loc[row_acq,'acquired_object_id'] == df_people.loc[row_people ,'relationship_object_id']:
            rows_list.append(
                get_row_dict(
                    df_acquisitions.loc[row_acq, 'acquisition_id'],
                    df_acquisitions.loc[row_acq, 'acquiring_object_id'],
                    df_acquisitions.loc[row_acq, 'acquired_object_id'],
                    df_acquisitions.loc[row_acq, 'term_code'],
                    df_acquisitions.loc[row_acq, 'price_amount'],
                    df_acquisitions.loc[row_acq, 'price_currency_code'],
                    df_acquisitions.loc[row_acq, 'acquired_at'],
                    df_acquisitions.loc[row_acq, 'source_url'],
                    df_acquisitions.loc[row_acq, 'source_description'],
                    df_acquisitions.loc[row_acq, 'created_at'],
                    df_acquisitions.loc[row_acq, 'updated_at'],
                    df_people.loc[row_people, 'full_name'],
                    df_people.loc[row_people, 'subject'],
                    df_people.loc[row_people, 'object_id'],  
                    True          
                )
            )
            
           # print('acqured ' + str(row_acq) + ' ' + str(row_people))
```
